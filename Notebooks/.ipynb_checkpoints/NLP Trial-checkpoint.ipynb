## Stopwords:
Words that are not essential to the semantic of the sentence

```python
#NLTK stopwords are a list:
nltk_stopwords = stopwords.words("english")
nltk_stopwords
```

```python
#Spacy stopwords. This is a set
STOP_WORDS
```

```python
#Creating a union of both stop words
stopwords_all = set(nltk_stopwords).union(STOP_WORDS)
```

```python
review = "This isn't a very good movie"
tokens = tokenizer.tokenize(review)
[token for token in tokens if token not in stopwords_all] #This is a problem!
```

```python
#Time to make an exception of stopwords:

exceptionStopWords = {
    'again',
    'against',
    'ain',
    'almost',
    'among',
    'amongst',
    'amount',
    'anyhow',
    'anyway',
    'aren',
    "aren't",
    'below',
    'bottom',
    'but',
    'cannot',
    'couldn',
    "couldn't",
    'didn',
    "didn't",
    'doesn',
    "doesn't",
    'don',
    "don't",
    'done',
    'down',
    'except',
    'few',
    'hadn',
    "hadn't",
    'hasn',
    "hasn't",
    'haven',
    "haven't",
    'however',
    'isn',
    "isn't",
    'least',
    'mightn',
    "mightn't",
    'must',
    'mustn',
    "mustn't",
    'needn',
    "needn't",
    'neither',
    'never',
    'nevertheless',
    'no',
    'nobody',
    'none',
    'noone',
    'nor',
    'not',
    'nothing',
    'should',
    "should've",
    'shouldn',
    "shouldn't",
    'too',
    'top',
    'up',
    'wasn',
    "wasn't",
    'well',
    'weren',
    "weren't",
    'won',
    "won't",
    'wouldn',
    "wouldn't",
}
#Remove exceptions from the complete set
final_stopwords = stopwords_all - exceptionStopWords
```

```python
review = "This isn't a very good movie"
tokens = tokenizer.tokenize(review)
[token for token in tokens if token not in final_stopwords] #This is a great!
```

```python
def remove_stopwords(reviews):
    return [token for token in reviews if token not in final_stopwords]
```

```python
reviews_new = list(map(lambda x : remove_stopwords(x),reviews_new))
```

## Lemmatization
Lemma of the word. i.e. basic form

```python
nlp= spacy.load("en", disable=["parser", "tagger","ner"])
```

```python
doc = nlp("isn't")
for token in doc:
    print(token.lemma_)
    
#is --> be
#n't --> not
```

```python
def lemmatization(review):
    lemma_result = []
    for words in review:
        doc = nlp(words) 
        for token in doc:
            lemma_result.append(token.lemma_)
    
    return lemma_result
```

```python
lemmatization(["this" , "isn't", "bad"])
```

```python
lemmatization(["walked" , "walking", "walker"])
```

```python
reviews_new = list(map(lambda x : lemmatization(x), reviews_new))
```
