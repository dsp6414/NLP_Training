```python
import re
import string
import nltk
from nltk.tokenize import TweetTokenizer
```

```python
with open('microblog2011.txt', 'r', encoding='utf8') as f:
    text = f.read()
```

```python
print("The first 1000 characters: ")
text[:1000]
```

```python
string.punctuation
```

```python
def sent_tokenize(text):
    text = re.split(r' *[\.\?!][\'"\)\]]* *', text)
    return text
```

```python
def word_tokenize(sent):
    translator = str.maketrans('', '', string.punctuation)
    word = sent.translate(translator)
    word = word.lower()
    word = re.split(r'\W+', word) 
    return word
```

```python
def count_words(text):
    counts = dict()
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator).lower()
    text = re.split(r'\W+', text) 
    for word in text:
        counts[word] = counts.get(word, 0) + 1
    return counts
```

```python
# question (a)
''' 
    sentence tokenization and word tokenization of the corpus
    output file: microblog2011_tokenized.txt
'''
nltk_sent_tokens = nltk.tokenize.sent_tokenize(text)
nltk_word_tokens = nltk.tokenize.word_tokenize(text)

with open("microblog2011_tokenized.txt", "a", encoding='utf8') as f:
        f.write(str(nltk_word_tokens))
        print("\nCorpus tokenized, output saved in microblog2011_tokenized.txt")
```

```python
# tokenization of the first 20 sentences
sents = ""
for sent in nltk_sent_tokens[:20]:
    sents += sent

nltk.tokenize.word_tokenize(sents)
```

```python
def test_run(text):
    counts = count_words(text)
    sorted_counts = sorted(counts.items(), key=lambda pair: pair[1], reverse=True)
        
    print("10 most common words:\nWord\tCount")
    for word, count in sorted_counts[:10]:
        print("{}\t{}".format(word, count))
        
    print("\n10 least common words:\nWord\tCount")
    for word, count in sorted_counts[-10:]:
        print("{}\t{}".format(word, count))
```

```python
test_run(text)
```

```python
# question (b)
token_counter = count_words(str(nltk_sent_tokens))

total_count = 0
for key in token_counter:
    total_count += token_counter[key]

print("Number of words (tokens): ", total_count)
print("Number of unique tokens (types): ", len(token_counter))
print("type/token ratio: ", len(token_counter) / total_count)
```

```python
# question (c)
sorted_types = {k: v for k, v in sorted(token_counter.items(), key=lambda item: item[1], reverse=True)}

with open("Tokens.txt", "a", encoding='utf8') as f:
    for key in sorted_types:
        output_string = ""
        output_string += key + ": " + str(sorted_types[key]) + "\n"
        f.write(output_string)
    print("Tokens and their frequency saved in Tokens.txt")
```

```python
with open('Tokens.txt', 'r', encoding='utf-8') as f:
    output = f.read().split("\n")[:100]
    
print("100 most frequent: ")
output
```

```python
# question (d)
appeared_once = {k for k in sorted_types if sorted_types[k] < 2}
appeared_once
print(len(appeared_once), "words appeared only once!")
```

```python
# question (e)
twit_exclude_tknzr = TweetTokenizer(strip_handles=True).tokenize(text)
print("Excluding twitter handles: ", len(twit_exclude_tknzr), "words")
```

```python
twit_counter = count_words(str(twit_exclude_tknzr))
sorted_twits = {k: v for k, v in sorted(twit_counter.items(), key=lambda item: item[1], reverse=True)}
sorted_twits
```

```python
total_twits = 0
for key in sorted_twits:
    total_twits += sorted_twits[key]

print("Number of words (tokens): ", total_twits)
print("Number of unique tokens (types): ", len(sorted_twits))
print("type/token ratio: ", len(sorted_twits) / total_twits)
```

```python
test_run(str(twit_exclude_tknzr))
```

```python
# question (f)
with open('stopwords.txt', 'r') as f:
    stopwords = f.read().split()
```

```python
stopped_words = [word for word in nltk_word_tokens if word not in stopwords]
count_stopped_words = count_words(str(stopped_words))
sorted_stopped_words = {k: v for k, v in sorted(count_stopped_words.items(), 
                                                key=lambda item: item[1], 
                                                reverse=True)}

total_stopped = 0
for key in sorted_stopped_words:
    total_stopped += sorted_stopped_words[key]
    
print("Number of words (tokens): ", total_stopped)
print("Number of unique tokens (types): ", len(sorted_stopped_words))
print("type/token ratio: ", len(sorted_stopped_words) / total_stopped)
```

```python
# question (g)
from nltk import bigrams

bigram = list(bigrams(count_stopped_words))

pair_counts = dict()
for pair in bigram:
    pair_counts[pair] = pair_counts.get(pair, 0) + 1 
    
sorted_pair_counts = {k: v for k, v in sorted(pair_counts.items(), key=lambda item: item[1], reverse=True)}

sorted_pair_counts
```
